### üîé System-Level Verification

```python
def count_lines(filename):
    filename = f"../Project-HDB-Store/{filename}"
    with open(filename, "r", encoding="utf-8") as f:
        return sum(1 for _ in f) - 1

total = 0
for child_id in downloader.child_datasets:
    file = f'DataLake/hdb_data_{child_id}.csv'
    num_lines = count_lines(file)
    total += num_lines
    print(f"{file} has {num_lines:7} lines.")
print(f"Total Records: {total}")
```
