```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# --- Sample time-series data ---
dates = pd.date_range(start="2024-01-01", periods=12, freq="M")
sales = np.random.randint(100, 300, size=12)      # Left axis
customers = np.random.randint(1000, 5000, size=12) # Right axis

# Create a DataFrame
df = pd.DataFrame({
    "date": dates,
    "sales": sales,
    "customers": customers
})

# --- Plot with two y-axes ---
fig, ax1 = plt.subplots(figsize=(10, 5))

# Left axis
ax1.plot(df["date"], df["sales"], marker="o", linewidth=2)
ax1.set_xlabel("Month")
ax1.set_ylabel("Sales (Units)", color="blue")
ax1.tick_params(axis='y', labelcolor='blue')

# Right axis
ax2 = ax1.twinx()  # <- Creates second y-axis
ax2.plot(df["date"], df["customers"], marker="s", linestyle="--")
ax2.set_ylabel("Customers Count", color="red")
ax2.tick_params(axis='y', labelcolor='red')

plt.grid()

plt.title("Sales vs Customers Over Time (Two Y-Axes)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# --- Create 12 months of data ---
dates = pd.date_range(start="2024-01-01", end="2024-12-31", freq="MS")
sales = np.random.randint(100, 300, size=12)
customers = np.random.randint(1000, 5000, size=12)

df = pd.DataFrame({
    "date": dates,
    "sales": sales,
    "customers": customers
})

# --- Plot ---
fig, ax1 = plt.subplots(figsize=(12, 5))

# Left y-axis
ax1.plot(df["date"], df["sales"], marker="o", linewidth=2, label="Sales")
ax1.set_ylabel("Sales", color="blue")
ax1.tick_params(axis='y', labelcolor='blue')

# Right y-axis
ax2 = ax1.twinx()
ax2.plot(df["date"], df["customers"], marker="s", linestyle="--",
         color="red", label="Customers")
ax2.set_ylabel("Customers", color="red")
ax2.tick_params(axis='y', labelcolor='red')

# --- X-axis ticks: show every month ---
plt.xticks(df["date"], df["date"].dt.strftime("%Y-%m"), rotation=45)

plt.grid()
plt.title("Sales vs Customers (One Full Year, Monthly)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# --- Generate sample monthly data ---
dates = pd.date_range(start="2024-01-01", periods=12, freq="MS")

# Create a clear peak for demonstration
sales = np.array([120, 150, 180, 200, 250, 300, 450, 280, 260, 220, 210, 190])
customers = np.random.randint(1500, 5000, size=12)

df = pd.DataFrame({
    "date": dates,
    "sales": sales,
    "customers": customers
})

# --- Find top peak ---
peak_idx = df["sales"].idxmax()
peak_date = df.loc[peak_idx, "date"]
peak_value = df.loc[peak_idx, "sales"]

# --- Plot ---
fig, ax1 = plt.subplots(figsize=(12, 5))

# Left y-axis (Sales)
ax1.plot(df["date"], df["sales"], marker="o", linewidth=2, label="Sales")
ax1.set_ylabel("Sales", color="blue")
ax1.tick_params(axis='y', labelcolor='blue')

# Highlight the peak month
ax1.scatter(peak_date, peak_value, color="red", s=120, zorder=5)
ax1.annotate(
    "Top Sales",
    xy=(peak_date, peak_value),
    xytext=(0, 15),
    textcoords="offset points",
    ha="center",
    color="red",
    fontsize=12,
    fontweight="bold"
)

# Right y-axis (Customers)
ax2 = ax1.twinx()
ax2.plot(df["date"], df["customers"], marker="s", linestyle="--", color="green", label="Customers")
ax2.set_ylabel("Customers", color="green")
ax2.tick_params(axis='y', labelcolor='green')

# X-axis ticks (every month)
plt.xticks(df["date"], df["date"].dt.strftime("%Y-%m"), rotation=45)

plt.title("Sales vs Customers (Highlight Top Peak Month)")
plt.tight_layout()
plt.show()
```
