# Using matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# --- Generate sample monthly data ---
dates = pd.date_range(start="2024-01-01", periods=12, freq="MS")

# Create a clear peak for demonstration
sales = np.array([120, 150, 180, 200, 250, 300, 750, 280, 260, 220, 210, 190])
customers = np.random.randint(1500, 5000, size=12)

df = pd.DataFrame({
    "date": dates,
    "sales": sales,
    "customers": customers
})

# --- Find top peak ---
peak_idx = df["sales"].idxmax()
peak_date = df.loc[peak_idx, "date"]
peak_value = df.loc[peak_idx, "sales"]

# --- Plot ---
fig, ax1 = plt.subplots(figsize=(12, 5))

# Left y-axis (Sales)
ax1.plot(df["date"], df["sales"], marker="o", linewidth=2, label="Sales")
ax1.set_ylabel("Sales", color="blue")
ax1.tick_params(axis='y', labelcolor='blue')

# Highlight the peak month
ax1.scatter(peak_date, peak_value, color="red", s=120, zorder=5)
ax1.annotate(
    "Top Sales",
    xy=(peak_date, peak_value),
    xytext=(0, -20),             #  move label BELOW the dot
    textcoords="offset points",
    ha="center",
    va="top",                    #  align text above its anchor
    color="red",
    fontsize=12,
    fontweight="bold"
)

# Right y-axis (Customers)
ax2 = ax1.twinx()
ax2.plot(df["date"], df["customers"], marker="s", linestyle="--",
         color="green", label="Customers")
ax2.set_ylabel("Customers", color="green")
ax2.tick_params(axis='y', labelcolor='green')

# X-axis ticks (every month)
plt.xticks(df["date"], df["date"].dt.strftime("%Y-%m"), rotation=45)

plt.grid()
plt.title("Sales vs Customers (Highlight Top Peak Month - Label Below Dot)")
plt.tight_layout()
plt.show()
```

# Usin Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# --- Generate sample monthly data ---
dates = pd.date_range(start="2024-01-01", periods=12, freq="MS")

# Create a clear peak for demonstration
sales = np.array([120, 150, 180, 200, 250, 300, 450, 280, 260, 220, 210, 190])
customers = np.random.randint(1500, 5000, size=12)

df = pd.DataFrame({
    "date": dates,
    "sales": sales,
    "customers": customers
})

# --- Find top peak ---
peak_idx = df["sales"].idxmax()
peak_date = df.loc[peak_idx, "date"]
peak_value = df.loc[peak_idx, "sales"]

# --- Seaborn styling ---
sns.set_style("whitegrid")

# --- Plot ---
fig, ax1 = plt.subplots(figsize=(12, 5))

# Left y-axis (Sales) using seaborn
sns.lineplot(data=df, x="date", y="sales",
             marker="o", linewidth=2, ax=ax1, color="blue", label="Sales")

ax1.set_ylabel("Sales", color="blue")
ax1.tick_params(axis='y', labelcolor='blue')

# Highlight the peak month
ax1.scatter(peak_date, peak_value, color="red", s=120, zorder=5)
ax1.annotate(
    "Top Sales",
    xy=(peak_date, peak_value),
    xytext=(0, -20),     # label BELOW dot
    textcoords="offset points",
    ha="center",
    va="top",
    color="red",
    fontsize=12,
    fontweight="bold"
)

# Right y-axis (Customers) using seaborn
ax2 = ax1.twinx()
sns.lineplot(data=df, x="date", y="customers",
             marker="s", linestyle="--", ax=ax2, color="green", label="Customers")

ax2.set_ylabel("Customers", color="green")
ax2.tick_params(axis='y', labelcolor='green')

# X-axis ticks (every month)
plt.xticks(df["date"], df["date"].dt.strftime("%Y-%m"), rotation=45)

plt.title("Seaborn: Sales vs Customers (Highlight Top Peak - Label Below Dot)")
plt.tight_layout()
plt.show()
```
